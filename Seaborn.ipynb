# SEABORN

```python
import matplotlib.pyplot as plt

# "matplotlib" is a --traditional AND manual-- way of plotting a graph in a python ,so we are using advanced one as "SEABORN" ....
```

```python
import pandas as pd
```

```python
dataset=pd.read_csv("pre_placement.csv")
dataset
```

```python
# for manually viewing a rows and columns....

print("number of rows:",dataset.shape[0])
print("number of column:",dataset.shape[1])
```

# Univariate_Analysis(SEABORN)

```python
import seaborn as sb
sb.displot(dataset["ssc_p"],kde=True) # it shows the frequency(repatation or count) and density(kde)....
```

# Bivariate_Analysis(SEABORN)

```python
# jointplot
```

```python
sb.jointplot(x="ssc_p",y="hsc_p",data=dataset) # joint plot will join 2 graphical method in one frame......
```

```python
sb.jointplot(x="ssc_p",y="hsc_p",data=dataset,kind="hex")
```

```python
sb.jointplot(x="ssc_p",y="hsc_p",data=dataset,kind="kde") # "kde"-kernel density estimation......
```

```python
# pairplot
```

```python
sb.pairplot(dataset,hue="gender",diag_kind="kde",kind="scatter",palette="husl") # diag_kind expose diagonal in "kde"
                                      # pairplot shows the relation and variance(differnce) of whole column in a dataset...            
```

# Catagorical_Plot(SEABORN)

```python
sb.stripplot(x="gender",y="ssc_p",data=dataset,jitter=True) # "jitter" improves the visual clarity od a plot....
```

```python
sb.swarmplot(x="gender",y="ssc_p",data=dataset) # swarmplot improves the visual clarity od a plot more than a stripplot.....
```

# DISTRIBUTION PLOT of catagorical .........."very important for analyizing"...

```python
sb.violinplot(x="specialisation",y="salary",data=dataset) # it will shows the distribution of given catagorical....
```

```python
sb.violinplot(x="specialisation",y="salary",hue="gender",data=dataset) # it will shows the distribution of given catagorical....
```

```python
sb.boxplot(x="ssc_b",y="ssc_p",data=dataset,hue="gender") # boxplot = "percentile" concept (Q1,Q2,Q3,Q4) & outlier ..WKT..
```

```python
sb.barplot(x="ssc_b",y="ssc_p",data=dataset,hue="gender") # barplot shows the overall max and min of both catagorical....
```

```python
sb.countplot(x="ssc_b",data=dataset,hue="gender") # countplot shows the overall cout of both catagories...
```

```python
sb.pointplot(x="ssc_b",y="ssc_p",data=dataset,hue="gender")
                                       # pointplot used for comparision of both catagories mean....
```
